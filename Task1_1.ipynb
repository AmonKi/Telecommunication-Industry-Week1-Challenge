```python
# importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Week1_clean_data.csv')
df.head(5)
```

```python
# Aggregation per user xDR sessions
df.aggregate({"MSISDN/Number":['sum','min','max']})
```

```python
# Aggregation per user session duration
df.aggregate({"Dur. (ms)":['sum','min','mean','max']})
```

```python
# Aggregation per user the total download (DL) and upload (UL) data
df.aggregate({"Total UL (Bytes)":['sum'],"Total DL (Bytes)":['sum']})
```

```python
df_total_data_vol = df[["Social Media DL (Bytes)","Social Media UL (Bytes)","Google DL (Bytes)","Google UL (Bytes)","Email DL (Bytes)","Email UL (Bytes)","Youtube DL (Bytes)","Youtube UL (Bytes)","Netflix DL (Bytes)","Netflix UL (Bytes)","Gaming DL (Bytes)","Gaming UL (Bytes)","Other DL (Bytes)","Other UL (Bytes)"]].sum()
df_total_data_vol
```
